```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("bongbongs/NewMes-v6")
model = AutoModelForCausalLM.from_pretrained("bongbongs/NewMes-v6")

# Move model to GPU and use half-precision if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
if torch.cuda.is_available():
    model.half()

# Define your input prompt
food_type = ' and '.join(['pizza', 'bread', 'juice'])
deases_type = ' and '.join(['Diabetes', 'heart attak'])

prompt = f"""
Act as Doctor speake about 'is eating every kind of this {food_type} is ok or not suitable in dependent for people who has {deases_type}'? in easy words and small answer
.
"""

# Tokenize the input prompt
inputs = tokenizer(prompt, return_tensors="pt").to(device)

# Generate text using the model with optimized parameters
outputs = model.generate(
    inputs.input_ids,
    max_length=300,  # Reduce max_length
    num_return_sequences=1,
    temperature=0.1,  # Adjust temperature for faster convergence
    top_k=50,  # Top-k sampling
    top_p=0.9,  # Nucleus sampling
    early_stopping=True  # Stop early if end of sentence is likely reached
)

# Decode the output tokens to text
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

print(generated_text)
```

```python
substring_to_remove = "\nAct as Doctor speake about 'is eating every kind of this pizza and bread and juice is ok or not suitable in dependent for people who has Diabetes and heart attak'? in easy words and small answer\n.\n"

result = generated_text.replace(substring_to_remove, "")
```
